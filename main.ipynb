## 2. Now write a function that returns the largest element on a list

```python
# Implementation avoiding all python internal functions

def greatest(numbers: list[float]):
    max_number = None

    if numbers != []:
        max_number = numbers[0]
        
        for number in numbers:
            
            if max_number == None or number > max_number:
                max_number = number

    else:
        print("The list is empty!")
    
    return max_number
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    if n == 1:
        return n
    
    else:
        return n*factorial(n-1)
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... ðŸ§`

```python
def mode_counter(arr):
    counter_dict = {}

    for item in arr:
        try:
            counter_dict[item]+=1
        except:
            counter_dict[item]=1
    

    element_highest_freq = ""
    max_value = 0
    for key, value in counter_dict.items():
        if value>max_value:
            max_value = value
            element_highest_freq = key

    return element_highest_freq
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. ðŸ˜‰`

```python
def st_dev(list_sd):
    total = 0.0
    n_items = 0.0

    for item in list_sd:
        total+=item
        n_items+=1

    assert total == sum(list_sd)
    assert n_items == len(list_sd)
    
    mean = total/n_items

    var = 0
    for item in list_sd:
        var+=(item-mean)**2

    var = var/(n_items-1)    
    std = var**(0.5)

    return std

```

```python
# This will test your function 
test_stdev(st_dev)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
    alphabeth = "qwertyuiopasdfghjklzxcvbnm"

    temp_string = string.lower()

    for letter in alphabeth:
        if letter not in temp_string:
            return False
        
    return True
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

```python
# def sort_alpha(string):
    
#     words = string.split(",")
#     sorted_words = sorted(words)
#     sorted_string = ",".join(sorted_words)

#     return sorted_string

def sort_alpha(string):
    
    word= ""
    comma = ","
    words = []

    for letter in string:
        if letter != comma:
            word+= letter
        else:
            words.append(word)
            word = ""

    # Add last word
    words.append(word)

    assert words == string.split(",")

    sorted_words = sorted(words)

    sorted_string = ""
    for i, word in enumerate(sorted_words):
        if i<len(sorted_words)-1:
            sorted_string +=word + comma
        else:
            sorted_string +=word


    assert ",".join(sorted_words) == sorted_string

    return sorted_string
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):

    if len(password) < 8:
        return False
    
    requirements = ["n_lower_case","n_upper_case","n_numbers", "n_special_char"]
    n_requirements = len(requirements)

    requirements_dict = dict(zip(requirements, [0]*n_requirements))

    for char in password:
        
        if char.isdigit():
            requirements_dict["n_numbers"]+=1
        
        elif char.isupper():
            requirements_dict["n_upper_case"]+=1
        
        elif char.islower():
            requirements_dict["n_lower_case"] +=1

        else:
            requirements_dict["n_special_char"]+=1        

    
    checks_requirements = [0 if item==0 else 1 for item in requirements_dict.values()]

    is_strong = sum(checks_requirements)==n_requirements
    return is_strong
    
```

```python
# This will test your function 
test_pass(check_pass)
```
